```python
# Import everything here
import cv2
import easyocr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class NepaliOCR:
    def __init__(self, gpu=True):
        """
        Initialize OCR for Nepali text.
        Use `gpu=True` if you have an NVIDIA GPU.
        """
        self.reader = easyocr.Reader(['ne'], gpu=gpu)  # 'ne' = Nepali
    
    def extract_text(self, image_path, visualize=False):
        """
        Extract text from an image.
        Returns: List of text lines.
        """
        img = cv2.imread(image_path)
        if img is None:
            raise FileNotFoundError(f"Image not found at {image_path}")
            
        results = self.reader.readtext(img, paragraph=True)
    
        # Safely extract text
        extracted_text = []
        for result in results:
            if len(result) >= 2:
                text = result[1]  # text is always at index 1
                extracted_text.append(text)
            else:
                print(f"Unexpected result format: {result}")
    
        if visualize:
            self._visualize_results(img, results)
    
        return extracted_text

    
    def _visualize_results(self, img, results):
   
    
        plt.figure(figsize=(15, 15))
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    
        for result in results:
            if len(result) >= 2:
                bbox, text = result[0], result[1]
            else:
                print(f"Unexpected result format: {result}")
                continue
    
            # Draw bounding box
            top_left = tuple(map(int, bbox[0]))
            bottom_right = tuple(map(int, bbox[2]))
            cv2.rectangle(img, top_left, bottom_right, (0, 255, 0), 2)
            cv2.putText(img, text, top_left, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)
    
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()
```

```python
# Initialize OCR (GPU=False for CPU)
ocr = NepaliOCR(gpu=True)

# Replace with your image path!
image_path = "testimage.jpg"  

# Extract and visualize text
text = ocr.extract_text(image_path, visualize=True)

print("\nExtracted Nepali Text:")
for i, line in enumerate(text, 1):
    print(f"{i}. {line}")
```
